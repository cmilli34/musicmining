# Music Moods

### I listen to a lot of punk, metal, and rap music. I feel like these genres have many overlaps in terms of their themes and moods that they convey. I also feel these genres bring a certain spunk and anger that feels like a sense of release. In general, I was curious to see what the artists I listen to most have in common in terms of their moods and themes, and what this says about me as a person (if it does at all).

## Data Collection: 
### I went to allmusic.com and researched all of the artists I love and collected the moods in one file and the themes in another file. I did this until I exhausted all of my favorite groups. Then, I saved those as text files. With these text files, I am going to make a word cloud and sentiment analysis.

```python
%matplotlib notebook
```

```python
import matplotlib.pyplot as plt
import os
import numpy as np
from wordcloud import WordCloud
import pandas as pd

#defining a space variable to use for cleaning below
spaces = " "
```

```python
#read in my mood file
#cleaning after loading to eliminate \t and \n and replacing with a space

moods_file = open("music_moods2.txt", "r")
moods_raw = moods_file.read()
moods = moods_raw.replace("\t", spaces)
moods = moods.replace("\n", spaces)
```

```python
#Read in the theme file
#cleaning after loading to eliminate \t and \n and replacing with a space

themes_file = open("music_themes.txt", "r")
themes_raw = themes_file.read()
themes = themes_raw.replace("\t", spaces)
themes = themes.replace("\n", spaces)
#eliminating the word "hanging" because I have no idea what that means in this context
themes = themes.replace("Hanging", spaces)
```

## WordClouds

```python
wordcloud = WordCloud(background_color = "white").generate(moods)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(background_color = "white").generate(themes)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
